**2.(i) Feature Selection:** From the correlation matrix, we can see that some features are highly correlated with each other, such as Area and DFactor4, Perimeter and DFactor5, MajorAxisLength and Perimeter, etc.
This means that these features are redundant and may cause multicollinearity problems for some models.
Therefore, we can try to remove some of these features and see if it improves the model performance.
We can use a threshold of 0.9 to select the features that have a correlation coefficient above this value with any other feature.

```python
#We can use the following code to find these features and drop them from the data
#Defining a function to find highly correlated features
def find_correlated_features(data, threshold): 
    correlated_features = set() 
    corr_matrix = data.corr()  
    for i in range(len(corr_matrix.columns)):
        for j in range(i): 
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i] 
                correlated_features.add(colname) 
                return correlated_features
```

```python
#Applying the function to our data with a threshold of 0.9
correlated_features = find_correlated_features(data, 0.9)
correlated_features
```

```python
#Dropping these features from the data
data.drop(correlated_features, axis=1, inplace=True)
```

Splitting data in to target and features, then into train and test data

```python
#Splitting the data into features (X) and target (y)
X = data.drop('Class', axis=1) 
y = data['Class']
```

```python
#Scaling the features using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Splitting the data into train and test sets with 80:20 ratio
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

**3. Testing Classification Algorithms**

```python
#Defining a function to evaluate different classification models
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
def evaluate_model(model, X_train, X_test, y_train, y_test):
    #Fitting the model on the train set 
    model.fit(X_train, y_train) 
    #Predicting on the test set 
    y_pred = model.predict(X_test) 
    #Calculating the accuracy score
    acc = accuracy_score(y_test, y_pred)
    #Printing the accuracy score 
    print(f"Accuracy: {acc:.4f}")
    #Printing the confusion matrix 
    print('Confusion Matrix:')
    print(confusion_matrix(y_test, y_pred))
    #Printing the classification report 
    print('Classification Report:') 
    print(classification_report(y_test, y_pred))
    #Calculating and printing the multiclass AUC score using ‘ovr’ method
    y_prob = model.predict_proba(X_test)
    auc = roc_auc_score(y_test, y_prob, multi_class='ovr')
    print(f"Multiclass AUC: {auc:.4f}")
```

```python
#Importing different models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Defining a list of models to compare
models = [LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier(), SVC(probability=True), KNeighborsClassifier()]
```

**4. Evaluating the predictions**

```python
#Looping through each model and evaluating it
for model in models:
    print(model)
    evaluate_model(model, X_train, X_test, y_train, y_test)
    print('-'*50)
```

**5. Comparing the methods:** We can see from the above result that **SVC** that is Support Vector Classifier performs best as it has highest accuracy that is 0.9332 and also highest multiclass AUC that is 0.9959 among all models. It also has the highest precision, recall, and f1-score for almost all the classes. Therefore, SVC is best algorithm fr this case.
